```python
from random import randint
from matplotlib import pyplot as plt
import numpy as np
```

```python
def showDadosNorma(jogadas):
  limite = jogadas*6
  resultados = np.zeros(limite)
  for i in range(99999):
    s=0
    for k in range(jogadas):
      s+=randint(1,6)
    resultados[s-1] += 1

  fig, ax = plt.subplots()
  ax.scatter(range(limite),resultados,color='b')
  plt.show()
```

```python
for i in range(5,50,5):
  showDadosNorma(i)
```
